```python
from bs4 import BeautifulSoup
import urllib
import re
import time
```

```python
url = "http://www.ikea.com/us/en/catalog/categories/departments/living_room/39130/"
```

```python
# html is simply a text file with markup
html = urllib.request.urlopen(url).read()

# see the text
print(html)
```

```python
# 'soup', is an object that allows fast querying for markup
soup = BeautifulSoup(html, "lxml")


titles = soup.select(".productTitle")

names = [title.text for title in titles]
print(names)
```

```python
names = set(names) # set removes duplicates

print("\n".join(names))
```

```python
# what if we also wanted to get the details?
details = soup.select(".productDetails > a")

print(details)
```

```python
links = [detail.attrs["href"] for detail in details]

# notice the links are not absolute links, they are relative
print(links)
```

```python
absolute_links = ["https://www.ikea.com" + link for link in links]

print(absolute_links)
```
